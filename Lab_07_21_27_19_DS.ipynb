```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"C:\Users\saura\Desktop\Ongoing\Notes\01.LAB_ass\Computer_vision_basics\Images\Lenna.png")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (5,10))
plt.imshow(img2)
```

##### Harris Corner Detector

```python
img_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
img_gray = np.float32(img_gray)
# CornerHarris (img, block size,k size, border size)
dest = cv2.cornerHarris(img_gray, 2, 5, 0.07)
dest = cv2.dilate(dest, None)
img2[dest > 0.01 * dest.max()] = [0, 0 , 255]

plt.figure(figsize = (5,10))
plt.imshow(dest , cmap = 'gray')
```
