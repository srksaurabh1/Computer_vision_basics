#### Import Dependencies

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

#### Read File using matplotlib

```python
image = plt.imread("Lenna.png")
plt.imshow(image)
print(f"Dimensions: {image.shape}")
# cv2.imshow("Lenna", image)
# cv2.waitKey(0)
```

We can use plt or cv2 for display of the image, but in jupyter lab cv2 will not display the image inline, and hence we required to add another line waitKey(0) to make it stay their until we close the window.

#### Read file using openCV package

```python
image2 = cv2.imread("Lenna.png")
plt.imshow(image2)
print(f"Dimensions: {image2.shape}")
# cv2.imshow("Lenna", image2)
# cv2.waitKey(0)
```

File imported using openCV reads the file to BGR format. Hence we need to convert it into RGB which is being used by most image compression tools nowadays.

#### Converting BGR -> RGB

```python
imageRGB = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(imageRGB)
print(f"Dimensions: {imageRGB.shape}")
```

#### Resizing Image

```python
width, height = 300,300
image_resize = cv2.resize(imageRGB, (width, height))
plt.imshow(image_resize)
print(f"Dimensions: {image_resize.shape}")
```

#### Converting Image to grayscale

```python
image_gray = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

We use cmap = 'gray' because matplotlib has separate colormap. 
cmap stands for colormap and it's a colormap instance or registered colormap name (cmap will only work if c is an array of floats). Matplotlib colormaps are divided into the following categories: sequential, diverging, and qualitative

#### Copying the image

#### Adding rectangle on image

```python
image_copy = np.copy(imageRGB)

start_point = (100, 100)
end_point = (400, 400)
color = (255, 0, 0)
thickness = 2

image_rect = cv2.rectangle(imageRGB, start_point, end_point, color, thickness)
plt.imshow(image_rect)
```

#### Crop Rectangle

```python
cropped_image = image_rect[100:400, 100:400]
plt.imshow(cropped_image)
```

#### Flip on Horizontal Axis, 0

```python
flip_on_Hori = cv2.flip(image_copy,0)
plt.imshow(flip_on_Hori)
```

#### Flip on vertical Axis, 1

```python
flip_on_Vert = cv2.flip(image_copy,1)
plt.imshow(flip_on_Vert)
```
