```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"C:\Users\saura\Desktop\Ongoing\Notes\01.LAB_ass\Computer_vision_basics\Images\bullet.jpg")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (5,10))
plt.imshow(img2)
```

##### Kernel

```python
kernel = np.ones((5,5))
```

##### Erosion

```python
image_erode = cv2.erode(img2, kernel)

plt.figure(figsize = (5,10))
plt.imshow(image_erode)
```

##### Dilation

```python
img_dilation = cv2.dilate(img2, kernel, iterations = 1)

plt.figure(figsize = (5,10))
plt.imshow(img_dilation)
```

##### Opening

```python
opening = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)

plt.figure(figsize = (5,10))
plt.imshow(opening)
```

##### Closing

```python
closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize = (5,10))
plt.imshow(image_erode)
```
