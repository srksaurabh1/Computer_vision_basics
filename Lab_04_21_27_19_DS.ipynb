Aim: To study and implement image edge detection techniques using python for processing image data

Edge Detection deals with the contours of an image usually denoted as an outline of a specific object in an image.

The goal of an edge detector is to enhance the connected gradients in an image, which may take the form
of an edge, contour, line, or some connected set of edges. Many edge detectors are simply implemented as
kernel operations, or convolutions
